```python
#since we have continous variables we are using boxplots to see outliers
plt.figure(figsize=(10,10))
for i, col in enumerate(data.columns):
    plt.subplot(3,3,i+1)
    plt.boxplot(data[col])
    plt.title(col)
    plt.ylabel(f"frequency")
plt.tight_layout() 
plt.show()   
```

we are having outliers in sr, and ddpi, the rest have no outliers.
